```python
import random

dinero = 100  # Saldo inicial del jugador
estado = "Abandonado"

print("Versión: OIIAOIIA")
print(f"Estado: {estado}")
print(f"Comienzas con un presupuesto de {dinero} puntos.")
print("Si te quedas sin dinero, mueres.")
print(".")
print(". Numeros: 1 entre 30")
print(". Colores: rojo, negro, verde")
print(".")
print("Let's go GAMBLING")

colores = ["Rojo", "Negro", "Verde"]
num = [str(n) for n in range(1, 31)]
par = [str(n) for n in range(2, 31, 2)]
impar = [str(n) for n in range(1, 30, 2)]
grupo1 = [str(n) for n in range(1, 11)]
grupo2 = [str(n) for n in range(11, 21)]
grupo3 = [str(n) for n in range(21, 31)]

def ruleta():
    global dinero
    try:
        apuesta = int(input("Cuánto quieres apostar? "))
        if apuesta > dinero:
            print("No tienes suficientes puntos.")
            return
    except ValueError:
        print("Valor insertado inválido.")
        return

    tnum = input("Número (1-30): ")
    tcolor = input("Color (rojo, negro, verde): ").capitalize()
    if tnum in par:
        ttipo = "par"
    else:
        ttipo = "impar"

    if tnum in grupo1:
        tgrupo = "grupo 1"
    elif tnum in grupo2:
        tgrupo = "grupo 2"
    else:
        tgrupo = "grupo 3"

    print(f"Has apostado por el número {tnum}, color {tcolor}, {ttipo} y {tgrupo}")

    # Resultados de la ruleta
    xnum = random.choice(num)
    xcolor = random.choice(colores)

    tipo = "par" if xnum in par else "impar"
    grupo = "grupo 1" if xnum in grupo1 else "grupo 2" if xnum in grupo2 else "grupo 3"
    color = xcolor.lower()

    print(f"El número ganador ha sido {xnum}, {tipo}, del {grupo} y de color {color}")

    # Multiplicadores de ganancias
    anum, acolor, agrupo, aparell = 10, 3, 3, 1.5
    apnume = anum if xnum == tnum else 1
    apcolor = acolor if color == tcolor.lower() else 1
    apgrupo = agrupo if grupo == tgrupo else 1
    aptipo = aparell if tipo == ttipo else 1

    ganancias = int(apuesta * apnume * apcolor * apgrupo * aptipo)
    dinero += ganancias - apuesta

    print(f"En total has ganado {ganancias} puntos!")
    print(f"Saldo actual: {dinero} puntos.")

while 0 < dinero < 10000:
    ruleta()

if dinero >= 10000:
    print("¡Felicidades! Has alcanzado más de 10,000 puntos y has ganado el juego.")
else:
    print("Te has quedado sin dinero, has perdido el juego.")
```
