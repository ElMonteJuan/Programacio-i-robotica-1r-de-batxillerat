```python
import random

# sustituir los def por elcodigo en si

hpmax = 0
atk = 0
crit = 0
critd = 0
defe = 0
dd = 0
emax = 0

enemigo_hpmax = 0
enemigo_atk = 0
enemigo_crit = 0
enemigo_critd = 0
enemigo_defe = 0
enemigo_dd = 0
enemigo_emax = 0

capa_hab_atk = 0
capa_hab_defe = 0
capa_hab_all = 0

enemigo_capa_hab_atk = 0
enemigo_capa_hab_defe = 0
enemigo_capa_hab_all = 0

daño_a_jugador = 0
daño_a_enemigo = 0

capa_curacion = 0
enemigo_capa_curacion = 0

damagesp = 0
enemigo_damagesp = 0


#___________________________________________________________________________JUGADOR

#ESTATUS DE LOS PERSONAJES
def personaje_hp():
  global hpmax, atk, crit, critd, defe, dd, emax
  hpmax = 158.0
  atk = 0.55
  crit = 0.2
  critd = 1.3
  defe = 0.31
  dd = 0.05
  emax = 70

def personaje_atk():
  global hpmax, atk, crit, critd, defe, dd, emax
  hpmax = 98.0
  atk = 0.91
  crit = 0.1
  critd = 1.6
  defe = 0.49
  dd = 0.07
  emax = 55

def personaje_cc():
  global hpmax, atk, crit, critd, defe, dd, emax
  hpmax = 93.0
  atk = 0.72
  crit = 0.35
  critd = 2.0
  defe = 0.39
  dd = 0.1
  emax = 90

def personaje_defe():
  global hpmax, atk, crit, critd, defe, dd, emax
  hpmax = 56.0
  atk = 0.54
  crit = 0.2
  critd = 1.3
  defe = 0.78
  dd = 0.03
  emax = 80

def personaje_all():
  global hpmax, atk, crit, critd, defe, dd, emax
  hpmax = 106.0
  atk = 0.67
  crit = 0.15
  critd = 1.5
  defe = 0.49
  dd = 0.07
  emax = 130


#PASIVAS DE LOS PERSONAJES
def habilidad_hp():
  global diferencia_hp, atk
  diferencia_hp = hpmax - hp
  atk = 0.55 + ((diferencia_hp/2.5)/100)


def habilidad_atk():
  global capa_de_atk_atk, capa_hab_atk, atk
  capa_de_atk_atk = 0.15
  capa_hab_atk = min(capa_hab_atk, 2)
  incremento_atk_atk = capa_de_atk_atk * capa_hab_atk
  atk = 0.91 + float(incremento_atk_atk)

def habilidad_defe():
  global capa_de_def_defe, capa_hab_defe, daño_espinas, defe
  capa_de_def_defe = 0.04
  capa_hab_defe = min(capa_hab_defe, 3)

  defe += capa_de_def_defe * capa_hab_defe
  daño_espinas = (defe * 0.08)


def habilidad_cc():
  global curacion_crit, daño_a_enemigo

  curacion_crit = (daño_a_enemigo * 0.5)
  hp += curacion_crit

def habilidad_all():
  global capa_de_hp_all, capa_hab_all, capa_de_atk_all, capa_de_crit_all, capa_de_critd_all, capa_de_defe_all, capa_de_dd_all
  capa_de_hp_all = 5
  capa_de_atk_all = 0.03
  capa_de_crit_all = 0.01
  capa_de_critd_all = 0.05
  capa_de_defe_all = 0.01
  capa_de_dd_all = 0.01
  capa_hab_all = min(capa_hab_all, 10)

  hp += capa_de_hp_all * capa_hab_all
  atk += capa_de_atk_all * capa_hab_all
  crit += capa_de_crit_all * capa_hab_all
  critd += capa_de_critd_all * capa_hab_all
  defe += capa_de_defe_all * capa_hab_all
  dd += capa_de_dd_all * capa_hab_all


#ULTI DE LOS PERSONAJES
def daño_ulti_hp():
  global damagesp, diferencia_hp
  damagesp = (diferencia_hp * 1.3)
  return damagesp

def daño_ulti_atk():
  global damagesp, atk
  damagesp = atk + (40 * capa_hab_atk)
  capa_hab_atk = 0
  return damagesp

def daño_ulti_defe():
  global damagesp, capa_hab_defe, defe
  damagesp = (defe * 1.7)
  capa_hab_defe += 1
  return damagesp

def daño_ulti_cc():
  global damagesp, enemigo_hp
  damagesp = (enemigo_hp * 0.8)
  return damagesp

def daño_ulti_all():
  global damagesp, curasp
  damagesp = (25 * capa_hab_all)
  curasp = (0.03 * capa_hab_all)
  return damagesp


#___________________________________________________________________________JUGADOR





#___________________________________________________________________________ENEMIGO

#ESTATUS DE LOS PERSONAJES ENEMIGOS
def enemigo_character_hp():
  global enemigo_hpmax, enemigo_atk, enemigo_crit, enemigo_critd, enemigo_defe, enemigo_dd, enemigo_emax
  enemigo_hpmax = 158.0
  enemigo_atk = 0.55
  enemigo_crit = 0.2
  enemigo_critd = 1.3
  enemigo_defe = 0.31
  enemigo_dd = 0.05
  enemigo_emax = 70

def enemigo_character_atk():
  global enemigo_hpmax, enemigo_atk, enemigo_crit, enemigo_critd, enemigo_defe, enemigo_dd, enemigo_emax
  enemigo_hpmax = 98.0
  enemigo_atk = 0.91
  enemigo_crit = 0.1
  enemigo_critd = 1.6
  enemigo_defe = 0.49
  enemigo_dd = 0.07
  enemigo_emax = 55

def enemigo_character_cc():
  global enemigo_hpmax, enemigo_atk, enemigo_crit, enemigo_critd, enemigo_defe, enemigo_dd, enemigo_emax
  enemigo_hpmax = 93.0
  enemigo_atk = 0.72
  enemigo_crit = 0.35
  enemigo_critd = 2.0
  enemigo_defe = 0.46
  enemigo_dd = 0.1
  enemigo_emax = 90

def enemigo_character_defe():
  global enemigo_hpmax, enemigo_atk, enemigo_crit, enemigo_critd, enemigo_defe, enemigo_dd, enemigo_emax
  enemigo_hpmax = 56.0
  enemigo_atk = 0.54
  enemigo_crit = 0.2
  enemigo_critd = 1.3
  enemigo_defe = 0.78
  enemigo_dd = 0.03
  enemigo_emax = 80

def enemigo_character_all():
  global enemigo_hpmax, enemigo_atk, enemigo_crit, enemigo_critd, enemigo_defe, enemigo_dd, enemigo_emax
  enemigo_hpmax = 106.0
  enemigo_atk = 0.67
  enemigo_crit = 0.15
  enemigo_critd = 1.5
  enemigo_defe = 0.49
  enemigo_dd = 0.07
  enemigo_emax = 130


#PASIVAS DE LOS PERSONAJES ENEMIGOS
def enemigo_habilidad_hp():
  global enemigo_diferencia_hp, enemigo_atk
  enemigo_diferencia_hp = enemigo_hpmax - enemigo_hp
  enemigo_atk = 0.55 + ((enemigo_diferencia_hp/2.5)/100)


def enemigo_habilidad_atk():
  global enemigo_capa_de_atk_atk, enemigo_capa_hab_atk, enemigo_atk
  enemigo_capa_de_atk_atk = 0.15
  enemigo_capa_hab_atk = min(enemigo_capa_hab_atk, 2)

  enemigo_atk += enemigo_capa_de_atk_atk * enemigo_capa_hab_atk

def enemigo_habilidad_defe():
  global enemigo_capa_de_def_defe, enemigo_capa_hab_defe, enemigo_daño_espinas, enemigo_hp, enemigo_defe
  enemigo_capa_de_def_defe = 0.04
  enemigo_capa_hab_defe = min(enemigo_capa_hab_defe, 3)

  enemigo_defe += enemigo_capa_de_def_defe * enemigo_capa_hab_defe
  enemigo_daño_espinas = (enemigo_defe * 0.08)


def enemigo_habilidad_cc():
  global enemigo_curacion_crit, daño_a_jugador, enemigo_hp

  enemigo_curacion_crit = (daño_a_jugador * 0.5)
  enemigo_hp += enemigo_curacion_crit



def enemigo_habilidad_all():
  global enemigo_capa_de_hp_all, enemigo_capa_hab_all, enemigo_hp, enemigo_atk, enemigo_crit, enemigo_critd, enemigo_defe, enemigo_dd
  enemigo_capa_de_hp_all = 5
  enemigo_capa_de_atk_all = 0.03
  enemigo_capa_de_crit_all = 0.01
  enemigo_capa_de_critd_all = 0.05
  enemigo_capa_de_defe_all = 0.01
  enemigo_capa_de_dd_all = 0.01
  enemigo_capa_hab_all = min(enemigo_capa_hab_all, 10)

  enemigo_hp += enemigo_capa_de_hp_all * enemigo_capa_hab_all
  enemigo_atk += enemigo_capa_de_atk_all * enemigo_capa_hab_all
  enemigo_crit += enemigo_capa_de_crit_all * enemigo_capa_hab_all
  enemigo_critd += enemigo_capa_de_critd_all * enemigo_capa_hab_all
  enemigo_defe += enemigo_capa_de_defe_all * enemigo_capa_hab_all
  enemigo_dd += enemigo_capa_de_dd_all * enemigo_capa_hab_all


#ULTI DE LOS PERSONAJES ENEMIGOS
def enemigo_daño_ulti_hp():
  global enemigo_damagesp, enemigo_diferencia_hp
  enemigo_damagesp = (enemigo_diferencia_hp * 1.3)
  return enemigo_damagesp

def enemigo_daño_ulti_atk():
  global enemigo_damagesp
  enemigo_damagesp = enemigo_atk + (20 * enemigo_capa_hab_atk)
  enemigo_capa_hab_atk = 0
  return enemigo_damagesp

def enemigo_daño_ulti_defe():
  global enemigo_damagesp, enemigo_capa_hab_defe,enemigo_defe
  enemigo_damagesp = (enemigo_defe * 1.7)
  enemigo_capa_hab_defe += 1
  return enemigo_damagesp

def enemigo_daño_ulti_cc():
  global enemigo_damagesp
  enemigo_damagesp = (hp * 0.8)
  return enemigo_damagesp

def enemigo_daño_ulti_all():
  global enemigo_damagesp
  enemigo_damagesp = (25 * enemigo_capa_hab_all)
  enemigo_curasp = (0.03 * enemigo_capa_hab_all)
  return enemigo_damagesp


#___________________________________________________________________________ENEMIGO





#___________________________________________________________________________ATAQUES

def probabilidad_atacar(probatk):
  probabilidad = random.randint(0, 100)
  if probatk >= probabilidad:
    return True
  elif probatk < probabilidad:
    return False

def AtaqueF():
    if probabilidad_atacar(60):
      return 90
    else:
      print ("El ataque ha sido fallado!")
      return 0

def AtaqueP():
  return 50

def AtaqueS():
    global damagesp, enemigo_damagesp

    damagesp = 0
    enemigo_damagesp = 0

    if accion == 4:
        if character_choice == "1":
            damagesp = daño_ulti_hp()
        elif character_choice == "2":
            damagesp = daño_ulti_atk()
        elif character_choice == "3":
            damagesp = daño_ulti_defe()
        elif character_choice == "4":
            damagesp = daño_ulti_cc()
        elif character_choice == "5":
            damagesp = daño_ulti_all()
        return damagesp

    if enemigo_accion == 4:
        if enemigo_character == "1":
            enemigo_damagesp = enemigo_daño_ulti_hp()
        elif enemigo_character == "2":
            enemigo_damagesp = enemigo_daño_ulti_atk()
        elif enemigo_character == "3":
            enemigo_damagesp = enemigo_daño_ulti_defe()
        elif enemigo_character == "4":
            enemigo_damagesp = enemigo_daño_ulti_cc()
        elif enemigo_character == "5":
            enemigo_damagesp = enemigo_daño_ulti_all()
        return enemigo_damagesp

    return 0



def Cura(vida):
  global cantidad_curacion, capa_res_curacion, capa_curacion, curacion_total, curacion_hp, enemigo_cantidad_curacion, enemigo_capa_res_curacion, enemigo_capa_curacion, enemigo_curacion_total, enemigo_curacion_hp
  if vida == hp:
    cantidad_curacion = 0.35
    capa_res_curacion = 0.05
    capa_curacion = min(capa_curacion, 4)
    curacion_total = (cantidad_curacion - (capa_res_curacion * capa_curacion))
    curacion_hp = curacion_total * hpmax
    return curacion_hp

  elif vida == enemigo_hp:
    enemigo_cantidad_curacion = 0.35
    enemigo_capa_res_curacion = 0.05
    enemigo_capa_curacion = min(enemigo_capa_curacion, 4)
    enemigo_curacion_total = (enemigo_cantidad_curacion - (enemigo_capa_res_curacion * enemigo_capa_curacion))
    enemigo_curacion_hp = enemigo_curacion_total * enemigo_hpmax
    return enemigo_curacion_hp

def critico(probcrit):
  global probabilidad_crit, hizo_critico, enemigo_nrg, nrg
  probabilidad_crit = random.uniform(0.00, 1.00)
  if probcrit >= probabilidad_crit:
    print ("Golpe critico!")
    hizo_critico = True # solo para el personaje cc
    if probcrit == crit:
      nrg += 20
    elif probcrit == enemigo_crit:
      enemigo_nrg += 20
    else:
      basura += 1
    return critd
  elif probcrit < probabilidad_crit:
    return 1
    hizo_critico = False


def esquivar(probdd):
  global probabilidad_dd
  probabilidad_dd = random.uniform(0.00, 1.00)
  if probdd >= probabilidad_dd:
    print ("El ataque ha sido evadido!")
    return 0
  elif probdd < probabilidad_dd:
    return 1

#___________________________________________________________________________ATAQUES




#___________________________________________________________________________ACCION


def barra_accion():
  global basura
  if nrg == emax and "|4| [S. Ataque] " not in acciones:
    acciones.append("|4| [S. Ataque] ")
  elif nrg != emax:
    if "|4| [S. Ataque] " in acciones:
      acciones.remove("|4| [S. Ataque] ")
    elif "|4| [S. Ataque] " not in acciones:
      basura += 1

def realizar_accion():
    global daño_a_enemigo, hp, curacion, capa_curacion, nrg, emax
    daño_a_enemigo = 0
    if accion == "1":
        nrg += 35
        daño_base = AtaqueF()
        daño_a_enemigo = ((atk * daño_base) * critico(crit) * (1 - enemigo_defe)) * esquivar(enemigo_dd)
    elif accion == "2":
        nrg += 20
        daño_base = AtaqueP()
        daño_a_enemigo = ((atk * daño_base) * critico(crit) * (1 - enemigo_defe)) * esquivar(enemigo_dd)
    elif accion == "3":
        nrg += 10
        hp = min(hp, hpmax)
        curacion = Cura(hp)
        hp += curacion
        hp = min(hp, hpmax)
        capa_curacion += 1
        capa_curacion = min(capa_curacion, 4)
        capa_curacion = max(capa_curacion, 0)
    elif accion == "4" and nrg == emax:
        nrg = 0
        nrg += 15
        defensa_enemiga = 0 if character_choice == "4" else 1
        daño_base = AtaqueS()
        daño_a_enemigo = ((atk * float(daño_base)) * critico(crit) * (1 - (enemigo_defe * defensa_enemiga))) * esquivar(enemigo_dd)
    if curacion > 0 :
      print (f"Te has curado {curacion:.1f} puntos de salud!")
    else:
      print (f"Has hecho {daño_a_enemigo:.1f} puntos de daño!")
      capa_curacion -= 1
      capa_curacion = min(capa_curacion, 4)
      capa_curacion = max(capa_curacion, 0)
    curacion = 0
    return daño_a_enemigo


def activar_pasiva():
  global character_choice, capa_hab_atk, enemigo_hp, daño_espinas, hizo_critico, basura, capa_hab_all
  if character_choice == "2" and daño_a_enemigo > 0:
    capa_hab_atk += 1
  elif character_choice == "3":
    enemigo_hp -= daño_espinas
  elif character_choice == "4":
    if hizo_critico == True:
      habilidad_cc()
    else:
      basura += 1
  elif character_choice == "5":
    capa_hab_all += 1


#___________________________________________________________________________ACCION





#___________________________________________________________________________ACCION_ENEMIGA

def enemigo_elegir_accion():
    global enemigo_accion, enemigo_nrg, enemigo_emax
    enemigo_accion = "1"
    if enemigo_nrg == enemigo_emax:
      enemigo_accion = "4"
    elif enemigo_nrg != enemigo_emax:
      if enemigo_hp > (enemigo_hpmax / 1.7):
        if hp >= (hpmax / 2.5):
            enemigo_accion = "1"
        elif hp < (hpmax / 2.5):
            enemigo_accion = "2"
      elif enemigo_hp <= (enemigo_hpmax / 1.7):
        enemigo_accion = "3"
    return enemigo_accion


def enemigo_realizar_accion():
    global daño_a_jugador, enemigo_hp, enemigo_curacion, enemigo_capa_curacion, enemigo_nrg, enemigo_emax
    daño_a_jugador = 0
    enemigo_hp = min(enemigo_hp, enemigo_hpmax)

    if enemigo_accion == "1":
        enemigo_nrg += 35
        enemigo_daño_base = AtaqueF()
        daño_a_jugador = ((enemigo_atk * enemigo_daño_base) * critico(enemigo_crit) * (1 - defe)) * esquivar(dd)
    elif enemigo_accion == "2":
        enemigo_nrg += 20
        enemigo_daño_base = AtaqueP()
        daño_a_jugador = ((enemigo_atk * enemigo_daño_base) * critico(enemigo_crit) * (1 - defe)) * esquivar(dd)
    elif enemigo_accion == "3":
        enemigo_nrg += 10
        enemigo_curacion = Cura(enemigo_hp)
        enemigo_hp += enemigo_curacion
        enemigo_hp = min(enemigo_hp, enemigo_hpmax)
        enemigo_capa_curacion += 1
        enemigo_capa_curacion = min(enemigo_capa_curacion, 4)
        enemigo_capa_curacion = max(enemigo_capa_curacion, 0)
    elif enemigo_accion == "4" and enemigo_nrg == enemigo_emax:
        enemigo_nrg = 0
        enemigo_nrg += 15
        defensa_jugador = 0 if enemigo_character == "4" else 1
        enemigo_daño_base = AtaqueS()
        daño_a_jugador = ((enemigo_atk * float(enemigo_daño_base)) * critico(enemigo_crit) * (1 - (defe * defensa_jugador))) * esquivar(dd)
    if enemigo_curacion > 0 :
      print (f"Se ha curado {enemigo_curacion:.1f} puntos de salud!")
    else:
      print (f"Te han hecho {daño_a_jugador:.1f} puntos de daño!")
      enemigo_capa_curacion -= 1
      enemigo_capa_curacion = min(enemigo_capa_curacion, 4)
      enemigo_capa_curacion = max(enemigo_capa_curacion, 0)
    enemigo_curacion = 0
    return daño_a_jugador

def enemigo_pasiva():
  if enemigo_character == "1":
    enemigo_habilidad_hp()
  elif enemigo_character == "2":
    enemigo_habilidad_atk()
  elif enemigo_character == "3":
    enemigo_habilidad_defe()
  elif enemigo_character == "4":
    enemigo_habilidad_cc()
  elif enemigo_character == "5":
    enemigo_habilidad_all()

def enemigo_activar_pasiva():
  global enemigo_character, enemigo_capa_hab_atk, hp, enemigo_daño_espinas, hizo_critico, basura, enemigo_capa_hab_all
  if enemigo_character == "2" and daño_a_jugador > 0:
    enemigo_capa_hab_atk += 1
  elif enemigo_character == "3":
    hp -= enemigo_daño_espinas
  elif enemigo_character == "4":
    if hizo_critico == True:
      enemigo_habilidad_cc()
    else:
      basura += 1
  elif enemigo_character == "5":
    enemigo_capa_hab_all += 1


#___________________________________________________________________________ACCION_ENEMIGA




#___________________________________________________________________________JUEGO



lista_personajes = ["1", "2", "3", "4", "5"]
print ("Elije tu personaje!")
character_choice = input("")
if character_choice in lista_personajes:
  lista_personajes.remove(character_choice)
  if character_choice == "1":
    personaje_hp()
  elif character_choice == "2":
    personaje_atk()
  elif character_choice == "3":
    personaje_defe()
  elif character_choice == "4":
    personaje_cc()
  elif character_choice == "5":
    personaje_all()

enemigo_character = random.choice(lista_personajes)
if enemigo_character in lista_personajes:
  if enemigo_character == "1":
    enemigo_character_hp()
  elif enemigo_character == "2":
    enemigo_character_atk()
  elif enemigo_character == "3":
    enemigo_character_defe()
  elif enemigo_character == "4":
    enemigo_character_cc()
  elif enemigo_character == "5":
    enemigo_character_all()

print ("QUE COMIENZE LA BATALLA!")
enemigo_hp = enemigo_hpmax
hp = hpmax
nrg = 0
nrg = min(nrg, emax)
enemigo_nrg = 0
enemigo_nrg = min(enemigo_nrg, enemigo_emax)
acciones = ["|1| [AtaqueF]",    "|2| [AtaqueP]",    "|3| [Curar]" ,    ]
basura = 0
curacion = 0
curacion_hp = 0
enemigo_curacion = 0
capa_curacion = 0
enemigo_capa_curacion = 0
daño_espinas = 0
enemigo_daño_espinas = 0


def revisar_pasivas():
    capa_hab_atk = min(capa_hab_atk, 2)
    capa_hab_defe = min(capa_hab_defe, 3)
    capa_hab_all = min(capa_hab_all, 10)

    enemigo_capa_hab_atk = min(enemigo_capa_hab_atk, 2)
    enemigo_capa_hab_defe = min(enemigo_capa_hab_defe, 3)
    enemigo_capa_hab_all = min(enemigo_capa_hab_all, 10)


#___________________________________________________________________________CORRER JUEGO

while enemigo_hp > 0 and hp > 0:
  nrg = min(nrg, emax)
  enemigo_nrg = min(enemigo_nrg, enemigo_emax)

  hp = min(hp, hpmax)
  enemigo_hp = min(enemigo_hp, enemigo_hpmax)

  revisar_pasivas()

  pasiva()
  enemigo_pasiva()

  print(f"JUGADOR: [hp:{hp:.1f}]  [atk:{(atk*100):.1f}]  [def:{(defe*100):.1f}%]  [crit:{(crit*100):.1f}%]  [critd:{(critd*100):.1f}%]  [dd:{(dd*100):.1f}%]  [nrg:{nrg}/{emax}]  ")
  print(f"ENEMIGO: [hp:{enemigo_hp:.1f}]  [atk:{(enemigo_atk*100):.1f}]  [def:{(enemigo_defe*100):.1f}%]  [crit:{(enemigo_crit*100):.1f}%]  [critd:{(enemigo_critd*100):.1f}%]  [dd:{(enemigo_dd*100):.1f}%]  [nrg:{enemigo_nrg}/{enemigo_emax}]  ")
  print ("Turno del JUGADOR")

  barra_accion()

  print (acciones)
  accion = input("")
  enemigo_hp -= float(realizar_accion())
  activar_pasiva()
  hizo_critico = False
  daño_a_enemigo = 0

  if enemigo_hp <= 0:
    print ("Has ganado!")
    break

  print ("Turno del ENEMIGO")

  enemigo_elegir_accion()
  hp -= float(enemigo_realizar_accion())
  enemigo_activar_pasiva()
  hizo_critico = False
  daño_a_jugador = 0

  if hp <= 0:
    print ("Has Perdido!")
    break
```
